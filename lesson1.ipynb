Bài 1: Tính cước taxi

```python
km = float(input("Nhập số km đã đi: "))
gia = 12000*km if km <= 10 else 12*10000 + 10000*(km-10)
gia = gia*0.9 if gia > 200000 else gia

print(f"{gia:,.0f} đồng")
print(f"Khuyến mãi: giảm 10% ({gia*100/90*0.1:,.0f} đồng)")
```

Bài 2: Quản lý thư viện sách

```python
sach = {
	"Python cơ bản": 3,
	"Phân tích dữ liệu": 2,
	"Trí tuệ nhân tạo": 1
}

sachMuon = input("Nhập tên sách muốn mượn: ")
if sachMuon in sach:
	if sach[sachMuon] > 0:
		sach[sachMuon] -= 1
		print("Mượn thành công.")
		print(f"Sách đã mượn: {sachMuon}")
	else:
		print("Sách đã hết.")
else:
	print("Sai tên sách")

print("Danh sách sách:")
for name, count in sach.items():
	print(f"- {name}: {count}")
```

Bài 3: Quản lý nhiệt độ

```python
listTemp = input("Nhập danh sách nhiệt độ (°C) trong 7 ngày: ")
listTemp = listTemp.split(" ")
listTemp = [float(i) for i in listTemp]
count = 0
for i in listTemp:
	if i > 30:
		count += 1

print(f"Nhiệt độ cao nhất: {max(listTemp)}°C")
print(f"Nhiệt độ thấp nhất: {min(listTemp)}°C")
print(f"Nhiệt độ trung bình: {sum(listTemp) / len(listTemp)}°C")
print(f"Số ngày có nhiệt độ trên 30°C: {count} ngày")
```

Bài 4: Kiểm tra mật khẩu mạnh

```python
def kiem_tra_mat_khau(pw: str) -> bool:
	if len(pw) < 8:
		return False
	cap = any(ch.isupper() for ch in pw)
	num = any(ch.isdigit() for ch in pw)
	return cap and num

pw = input("Nhập mật khẩu: ")
print("Mật khẩu mạnh" if kiem_tra_mat_khau(pw) else "Mật khẩu yếu")
```
